This notebook implements image restoration and object detection..

```python
import os
import cv2
import numpy as np
from skimage.io import imread
from skimage.color import rgb2gray
from skimage.restoration import wiener, richardson_lucy
import matplotlib.pyplot as plt
from skimage.metrics import peak_signal_noise_ratio, structural_similarity
```

```python
blur_folder = "../data/blur"
sharp_folder = "../data/sharp"
```

```python
files = os.listdir(blur_folder)
```

```python
# store result for each image and compute avg performance later
psnr_wiener_list = []
psnr_rl_list = []

ssim_wiener_list = []
ssim_rl_list = []
```

```python
# point spread function
psf = np.ones((5,5))/25
```

```python
for file in files:
    blur = cv2.imread(os.path.join(blur_folder, file[0]))
    sharp = cv2.imread(os.path.join(sharp_folder, file))

    blur = cv2.cvtColor(blur, cv2.COLOR_BGR2RGB)
    sharp = cv2.cvtColor(sharp, cv2.COLOR_BGR2RGB)

    blur_gray = rgb2gray(blur)
    sharp_gray = rgb2gray(sharp)

    wiener_img = wiener(blur_gray, psf, balance=0.1)
```

```python
plt.imshow(wiener_img, cmap='gray')
plt.title("Deblurred Image")
plt.axis("off")
plt.show()
```
